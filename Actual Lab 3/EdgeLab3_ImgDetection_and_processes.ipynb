```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error")
else:
    print("Webcam detected successfully")

cap.release()
```

```python
cap = cv2.VideoCapture(0)

ret, frame = cap.read()

if ret:
    
    h, w, _ = frame.shape
    
    cx = w // 2
    cy = h // 2
    
    B, G, R = frame[cy, cx]
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    gray_value = gray[cy, cx]
    
    print("Center Coordinates:", cx, cy)
    print("RGB Values:", (R, G, B))
    print("Grayscale Value:", gray_value)
    
    cv2.circle(frame, (cx,cy), 5, (0,255,0), -1)
    
    cv2.imshow("Captured Frame", frame)
    
    cv2.waitKey(0)

cap.release()
cv2.destroyAllWindows()
```

```python
#grayscale feed
```

```python
cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()
    
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    cv2.imshow("Live Grayscale Feed", gray)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#canny edge detection
```

```python
cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()
    
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    edges = cv2.Canny(gray,100,200)

    cv2.imshow("Live Canny Edge Detection", edges)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()
    
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    negative = 255 - gray

    cv2.imshow("Live Negative", negative)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
