```python
import pandas as pd
# The URL for the uncleaned dataset
url = 'raw.githubusercontent.com'

try:
    # Use pandas to read the CSV, specifying the separator and engine for better parsing
    df = pd.read_csv(url, sep=',', engine='python')
    print("Dataset loaded successfully into a DataFrame named 'df'.")
    print(f"Dataset shape: {df.shape}")
except Exception as e:
    print(f"An error occurred while loading the data: {e}")
```

```python
# Check the shape of the DataFrame (rows, columns)
print(f"Dataset shape: {df.shape}")
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
```

```python
# Display the first 5 rows of the dataset
print("First 5 rows of the DataFrame:")
display(df.head())
# Get a concise summary of the DataFrame
print("DataFrame information and data types:")
df.info()
```

```python
# Display descriptive statistics for numerical columns
print("Descriptive statistics for numerical columns:")
display(df.describe())
```
